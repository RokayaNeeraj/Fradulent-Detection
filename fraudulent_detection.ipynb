# Building Random Forest Model

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, confusion_matrix, plot_confusion_matrix
import matplotlib.pyplot as plt

# Splitting the data into features and target
X = df.drop('class', axis=1)
y = df['class']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Building a random forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Predictions and AUC
y_train_pred = rf_model.predict(X_train)
y_test_pred = rf_model.predict(X_test)
auc_train = roc_auc_score(y_train, y_train_pred)
auc_test = roc_auc_score(y_test, y_test_pred)

# Confusion Matrices
conf_matrix_train = confusion_matrix(y_train, y_train_pred)
conf_matrix_test = confusion_matrix(y_test, y_test_pred)

# Plotting confusion matrix
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
plot_confusion_matrix(rf_model, X_train, y_train, ax=ax[0], cmap='Blues')
ax[0].set_title('Confusion Matrix - Training Data')
plot_confusion_matrix(rf_model, X_test, y_test, ax=ax[1], cmap='Blues')
ax[1].set_title('Confusion Matrix - Test Data')
plt.tight_layout()

# Output the AUC and Confusion Matrix results
auc_train, auc_test, conf_matrix_train, conf_matrix_test, plt.show()
```

# Round 1 of tuning: Adjusting 'n_estimators' and 'max_depth'

```python
from sklearn.model_selection import GridSearchCV
param_grid_1 = {
    'n_estimators': [50, 100, 150],  # Number of trees
    'max_depth': [10, 20, 30]        # Maximum depth of each tree
}

# Grid search for the first round of tuning
grid_search_1 = GridSearchCV(estimator=RandomForestClassifier(random_state=42),
                             param_grid=param_grid_1,
                             cv=3,
                             scoring='roc_auc',
                             verbose=1)

grid_search_1.fit(X_train, y_train)

# Best parameters and score from the first round
best_params_1 = grid_search_1.best_params_
best_score_1 = grid_search_1.best_score_
best_params_1, best_score_1
```

# Round 2 of tuning: Adjusting 'min_samples_split' and 'min_samples_leaf'

```python
param_grid_2 = {
    'min_samples_split': [2, 4, 6],  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4]    # Minimum number of samples required to be at a leaf node
}

# Updating the model with the best parameters from round 1
rf_model_tuned = RandomForestClassifier(random_state=42,
                                        n_estimators=best_params_1['n_estimators'],
                                        max_depth=best_params_1['max_depth'])

# Grid search for the second round of tuning
grid_search_2 = GridSearchCV(estimator=rf_model_tuned,
                             param_grid=param_grid_2,
                             cv=3,
                             scoring='roc_auc',
                             verbose=1)

grid_search_2.fit(X_train, y_train)

# Best parameters and score from the second round
best_params_2 = grid_search_2.best_params_
best_score_2 = grid_search_2.best_score_

best_params_2, best_score_2
```

# Using permutation importance for interpretability

```python
perm_importance = permutation_importance(rf_model, X_test, y_test, n_repeats=10, random_state=42)

# Creating a series for the permutation importance
perm_importance_series = pd.Series(perm_importance.importances_mean, index=X.columns)

# Sorting and plotting
sorted_perm_importance = perm_importance_series.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=sorted_perm_importance, y=sorted_perm_importance.index)
plt.title("Permutation Importances on Test Set")
plt.xlabel("Permutation Importance")
plt.ylabel("Features")
plt.show()

# Output the top 5 features based on permutation importance
top_5_features_perm = sorted_perm_importance.head(5)
top_5_features_perm
```
